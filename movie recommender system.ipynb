```python
import numpy as np 
import pandas as pd
import ast as ast
```

## Data Preprocessing

```python
movies= pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
# merging two dataset based on title of movie  and storing in movies variable
movies= movies.merge(credits, on='title')
```

```python
movies.shape
```

```python
movies.head()
```

```python
## column selected 
#genres
#title
#cast
#crew
#id
#Kyeword
#overview

movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
# Overview as 3 missing or null data so drop them 
movies.dropna(inplace=True)
## now removed
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
 movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
## fetching only first 3 cast name form cast feature
def convert3(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count!=3:
            L.append(i['name'])
            count+=1
        else:     
            break
    return L        
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
## fetching only job= director from the crew feature

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break;
    return L        
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
## overview column is string and we will convert it into list 
## because all other columns like cast crew keywords we have converted it to list

movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

###Transformation

#Tranformation is done to remove the spaces from any columns in which cast name in Sam Worthingon so we have to remove space and make it SamWorthingon
We do This because in crew column also  we have name of crew as Sam Mendes so while training the model, the model may get confused and make error so remove the spaces and make it one string

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
## Now we will create the "tag" cloumn and concatenate all 5 columns:- 
## overview, genres, keywords, cast, crew

movies['tags'] = movies['overview'] + movies['keywords'] + movies['cast'] + movies['crew']

movies.head()
```

#Now the next step is to remove all the colums that we concatinated and keep only tag colums

#We will be creating new data frame:- new_df 

```python
new_df= movies[['movie_id', 'title', 'tags']]
new_df.head()
```

## Stemming

#Now this can happen that feature our words that we took as array can have same name features like actions and action in above cell( it is not showing here right now) now we just want only one name either action or actions. So we apply Stemming Technique

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y =[]
    for i in text.split():
        y.append(ps.stem(i))
    return  " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

# Euclidean and cosine Similarity 


When we want to find the distance between any two vectors both the type of distance is used

Euclidean distance find the distance betwn the vectors

Cosine similarity measures the similarity between two vectors of an inner product space. It is measured by the cosine of the angle between two vectors and determines whether two vectors are pointing in roughly the same direction. It is often used to measure document similarity in text analysis. if the angle is less the movies will be more similar and if angle in more like 180 degree then movies have different and opposite content 

We will be using cosine distance because Euclidean distance is not suitable for the high dimensional space and we have 5000 dimension. Therefore use cosine distance






```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

## Recommendation function
Now make recommend function and return 5 most similar movies based on user  input movie 

#Steps
1. find the index of user input movie 
2. Then go to similarity array and find its index Ex:- 
   if Avatar is movie then get array of similarity[0]
3. Then sort the array of that index movie like Avatar     has array which contains its cosine similarity from     other movies then sort the array and we get most       similar movie in front and less similar movie at       the back
4. Then fetch first 5 similar movies from sorted list       of arrays.

```python
#code to find index of any movie
new_df[new_df['title'] == 'Batman Begins'].index[0]
## index if Batman begins movie is 199    
    
```

```python
## sorting and at the same time retaining and getting index of the first 5 similar movies

sorted(list(enumerate(similarity[0])) , reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)) , reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
## calling recommend function and printing 5 similar movies
recommend('Avatar')
```
